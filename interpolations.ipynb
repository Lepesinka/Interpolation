# Interpolation examples
https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp1d.html#scipy.interpolate.interp1d

```python
import matplotlib.pyplot as plt
from scipy import interpolate
import numpy as np
```

```python


x = array([ 0, 20, 45, 53, 57, 62, 74, 89, 95,100])
y = array([47, 47,  0,382, 73,434,151, 47,147, 47])
```

# Final

```python
plt.figure(figsize=(15,8))
plt.plot(x, y, 'o', markersize=10, label='real points')
xnew = np.arange(0, 100, 0.5)

#pchip, TPS, multiquad

f = interpolate.PchipInterpolator(x, y)
ynew = f(xnew)
plt.plot(xnew, ynew, '-', label='PCHIP')

kernels = [
    'thin_plate_spline',
    'multiquadric',
]

for kernel in kernels:
    f = interpolate.RBFInterpolator(expand(x), y, kernel=kernel, epsilon=2)
    ynew = f(expand(xnew))
    plt.plot(xnew, ynew, '-', label='RBF ({})'.format(kernel))

plt.grid(True)
plt.legend(loc='best')
plt.ylim(-400, 800)
#plt.show()
plt.savefig('interp all.pdf')
```
